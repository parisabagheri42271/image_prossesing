```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = Image.open("/content/ex2.png")
```

```python
gray_img = img.convert("L")
```

```python
gray_array = np.array(gray_img)
```

```python
max_pixel = np.max(gray_array)
min_pixel = np.min(gray_array)

print("Maximum Pixel Value:", max_pixel)
print("Minimum Pixel Value:", min_pixel)
```

```python
MIN = 0
MAX = 255

# Perform the operation: (pixel - min) / (max - min)
result_array = ((gray_array - min_pixel) / (max_pixel - min_pixel)) *(MAX - MIN) + MIN
```

```python
# Convert back to an image for display
result_img = Image.fromarray(result_array)
gray_img = Image.fromarray(gray_array)
```

```python
# Display the result image
plt.imshow(img, cmap='gray')
plt.axis('off')  # Turn off axis
plt.show()
```

```python
result_histogram2 = gray_img.histogram()

# Plot the histogram
plt.figure(figsize=(10, 5))
plt.title("Histogram of Result Image")
plt.xlabel("Pixel Value")
plt.ylabel("Frequency")
plt.bar(range(256), result_histogram2, color='black')
plt.xlim(0, 255)
plt.show()
```

```python
result_histogram = result_img.histogram()

# Plot the histogram
plt.figure(figsize=(10, 5))
plt.title("Histogram of Result Image")
plt.xlabel("Pixel Value")
plt.ylabel("Frequency")
plt.bar(range(256), result_histogram, color='black')
plt.xlim(0, 255)
plt.show()
```

```python
# Display the result image
plt.imshow(result_img, cmap='gray')
plt.axis('off')  # Turn off axis
plt.show()
```
